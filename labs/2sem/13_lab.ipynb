```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from src.inout import InOuter
from src.model import Model
from src.analysis import Analyzer
from src.processing import Processor

inout = InOuter()
model = Model()
analyzer = Analyzer()
processor = Processor()
img_path = '../../data/'
```

```python
def contour_with_erosion(img, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.uint8) 
    er_img = cv2.erode(img.astype(np.uint8), kernel)
    return np.array(img - er_img, dtype=np.uint8)


def contour_with_dilation(img, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.uint8) 
    dil_img = cv2.dilate(img.astype(np.uint8), kernel)
    return  np.array(dil_img - img, dtype=np.uint8)
```

```python
img = inout.read_jpg(img_path+'MODELimage.jpg')

fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(img, 'gray')
ax[1].imshow(cv2.threshold(contour_with_erosion(img, 3), 100, 255, cv2.THRESH_BINARY)[1], 'gray')
ax[2].imshow(cv2.threshold(contour_with_dilation(img, 3), 100, 255, cv2.THRESH_BINARY)[1], 'gray')
```
