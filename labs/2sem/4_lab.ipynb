```python
import matplotlib.pyplot as plt
import numpy as np

from src.inout import InOuter
from src.model import Model
from src.analysis import Analyzer
from src.processing import Processor

inout = InOuter()
model = Model()
analyzer = Analyzer()
processor = Processor()
img_path = '../../data/'

def do_lab(img_name, gamma_const, gamma_gamma, log_const):
    fig, ax = plt.subplots(2, 2, figsize=(10, 10))
    img = inout.read_jpg(img_name)
    negative_img = processor.negative(img)
    log_img = np.clip(processor.log_correction(img, log_const), 0, 255).astype(np.uint8)
    gamma_img = np.clip(processor.gamma_correction(img, gamma_const, gamma_gamma), 0, 255).astype(np.uint8)
    
    ax[0, 0].imshow(img)
    ax[0, 1].imshow(negative_img)
    ax[1, 0].imshow(log_img)
    ax[1, 1].imshow(gamma_img)
    
    
    fig1, ax1 = plt.subplots(2, 2, figsize=(10, 10))
    ax1[0, 0].hist(img.ravel(),256,[0,256])
    ax1[0, 1].hist(negative_img.ravel(),256,[0,256])
    ax1[1, 0].hist(log_img.ravel(),256,[0,256])
    ax1[1, 1].hist(gamma_img.ravel(),256,[0,256])
```

```python
do_lab(img_path + 'photo1.jpg', 5, 0.66, 33)
```
